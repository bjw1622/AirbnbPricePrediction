```python
import seaborn as sns
heatmap_data = data[['Private room', 'Shared room', 'Latitude', 'Number of Reviews', 'Pull-out Sofa', 'Futon', 'Couch', 'Airbed', 'Maximum Nights', 'Minimum Nights',
                     'Longitude', 'Real Bed', 'Guests Included', 'Extra People', 'Bathrooms', 'Beds', 'Entire home/apt', 'Bedrooms', 'Accommodates', 'Cleaning Fee', 'Price']]

colormap = plt.cm.PuBu
plt.figure(figsize=(15, 10))
plt.title("Listing Correlation of Features", y = 1.05, size = 10)
sns.heatmap(heatmap_data.astype(float).corr(), linewidths = 0.1, vmax = 1.0,
            square = True, cmap = colormap, linecolor = 'white', annot = True, annot_kws = {"size" : 10})
```
